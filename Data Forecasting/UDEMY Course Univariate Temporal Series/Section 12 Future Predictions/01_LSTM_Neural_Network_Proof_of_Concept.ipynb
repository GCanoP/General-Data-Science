```python
# Imoporting Relevant Packages. 
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
from jupyterthemes import jtplot
jtplot.style(theme='monokai', context = 'notebook', ticks = True, grid = True)
```

```python
# Importing the Dataset / The Dataset is Clean no Preprocessing is Needed.all 
data = pd.read_csv('airline_passengers.csv')
data.Month = pd.to_datetime(data.Month)
data.set_index('Month', inplace = True)
data = data.asfreq('MS')
print(data.head())
print(data.describe())
```

```python
# Plotting the Dataset. 
data.plot(figsize = (24,6))
plt.title('Airline Passengers Dataset')
plt.ylabel('Thousnads of Passengers')
plt.grid(True)
plt.show()
```
