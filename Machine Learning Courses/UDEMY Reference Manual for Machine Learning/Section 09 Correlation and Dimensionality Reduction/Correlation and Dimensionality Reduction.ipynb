## <font color='cornflowerblue'> Correlation Analysis. </font>
___
Correlations in data can be:
- __Positive__ : Both variables increases. 
- __Negative__: One variable increases, while the other decreases. 
- __Nule__ : There is no relation between both variables. 

__(Note)__ Covariance is not a useful metric to compare two different variables as is not adimensional.

### <font color='teal'> Pearson Linear Correlation. </font>
___
__Linear Person Correlation__ : To avoid the problem of covariance, Person´s correlation coefficient is introduced as the division betweem the covariance and the standard deviation of variables. This coefficient measures the linear correlations. 

$$
r = \frac{S_{xy}}{S_x S_y} = \frac{Covariance}{Standard Deviation} = Person's\ Correlation
$$

(0 = no linear correlation, 1 = perfect linear correlation, -1 perfect inverse linear correlation).

- __Nule Hypothesis__: There is no correlation between variables.
- __Alternative Hypothesis__: There is a correlation between variables.

```python
# Ejemplo. 
X = [105, 116, 103, 124, 137, 126, 112, 129, 118, 105]
y = [4, 8, 2, 7, 9, 9, 3, 10, 7, 6]
```

```python
# Plot the data
plt.figure(figsize = (12, 4))
plt.scatter(X,y)
```

```python
# Compute the covariance matrix. 
covariance_matrix = np.cov(X, y)/(np.multiply(np.std(X, ddof=1), np.std(y, ddof=1)))
print('Covariance (X-y) = {covariance_xy}'.format(covariance_xy = covariance_matrix[0, 1]))
```

```python
# Compute t value. 
ro = covariance_matrix[0,1]

t = (ro)/np.sqrt((1- ro**2)/(len(X)-2))
print("t-computed = {t_value}".format(t_value = t))
print("t-theoretical = 2.306")
print("Nule Hypothesis is Rejected = Variables have Linear Correlation")
```

```python
# Using numpy functions. 
np.corrcoef(X,y)
```
